```python
# STEP 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For better visuals
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

```

```python
# ========================================
# IPL Deliveries Dataset - Exploratory Data Analysis

# STEP 2: Load Dataset
df = pd.read_csv(r"C:\Users\NINAD\Downloads\archive\deliveries.csv")

print("‚úÖ Dataset Loaded Successfully!")
print("Shape of DataFrame:", df.shape)

```

```python
# ========================================
# IPL Deliveries Dataset - Exploratory Data Analysis
# Author: [Your Name]
# Level: Fresher / Beginner
# ========================================


# Display first few rows
print(df.head())

# STEP 3: Basic Information
print("\n=== Dataset Info ===")
print(df.info())

print("\n=== Null Values ===")
print(df.isnull().sum())

```

```python


# STEP 4: Basic Stats
print("\n=== Basic Statistics ===")
print(df.describe())

# STEP 5: Top 5 Batsmen by Total Runs
top_batsmen = df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Batsmen by Total Runs:\n", top_batsmen)

sns.barplot(y=top_batsmen.index, x=top_batsmen.values, palette="viridis")
plt.title("Top 10 Batsmen by Total Runs")
plt.xlabel("Total Runs")
plt.ylabel("Batsman")
plt.show()

```

```python


# STEP 6: Top 5 Bowlers by Wickets
# A wicket is counted if 'player_dismissed' column is not null
top_bowlers = df[df['player_dismissed'].notna()]['bowler'].value_counts().head(10)
print("\nTop 10 Bowlers by Wickets:\n", top_bowlers)

sns.barplot(y=top_bowlers.index, x=top_bowlers.values, palette="rocket")
plt.title("Top 10 Bowlers by Wickets")
plt.xlabel("Wickets Taken")
plt.ylabel("Bowler")
plt.show()

# STEP 7: Most Runs by Batting Teams
team_runs = df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)
print("\nTotal Runs by Each Team:\n", team_runs)

sns.barplot(y=team_runs.index, x=team_runs.values, palette="cool")
plt.title("Total Runs Scored by Each Team")
plt.xlabel("Total Runs")
plt.ylabel("Team")
plt.show()


```
